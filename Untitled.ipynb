```python
import pandas as pd

# Read the CSV file into a DataFrame
df=pd.read_csv("D:\BE Final Year Project\workspace\Outputs\CSV\merged_df.csv")
df['checkpoint'] = (df['scorecard'] & df['bowler']).astype(int)
bowler_df = df[df['checkpoint'] == 1]
print(bowler_df)
# Calculate the difference between consecutive sec values
df['diff'] = df['sec'].diff()

# Group consecutive rows with a difference of one using a cumulative sum
df['group'] = (df['diff'] != 1).cumsum()
print(df)
# Aggregate the DataFrame by taking the minimum sec value within each group
merged_df = df.groupby('group').agg({'sec': 'min', 'scorecard': 'min', 'bowler': 'min', 'checkpoint': 'min'})

# Reset the index to remove the 'group' column and set the index back to sequential integers
merged_df.reset_index(drop=True, inplace=True)

# Print the merged DataFrame
print(merged_df)
```
